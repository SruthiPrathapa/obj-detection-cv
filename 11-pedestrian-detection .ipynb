```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
files = os.listdir("data/dataset")

files
```

```python
img_path_list = []

for f in files:
    
    if f.endswith(".jpg"):
        img_path_list.append(f)
        
img_path_list
```

# hog descriptor

```python
# define hog Descriptor
hog = cv2.HOGDescriptor()

# add SVM
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

for img_path in img_path_list:
    
    # print img directory
    print(f"data/dataset/{img_path}")
    
    # read images
    image = cv2.imread(f"data/dataset/{img_path}")
    
    # detect
    (rects, weights) = hog.detectMultiScale(image, padding=(8,8), scale=1.05)
    
    for (x,y,w,h) in rects:
        
        # draw rectangle
        cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255), 2)
        
    cv2.imshow("Pedestrian", image)
    
    if cv2.waitKey(0) & 0xFF == ord("q"): continue
        
cv2.destroyAllWindows()
```
