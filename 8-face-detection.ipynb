```python
import cv2
import matplotlib.pyplot as plt
```

```python
einstein = cv2.imread("data/einstein.jpg", 0)

plt.figure(), plt.title("Einstein"), plt.imshow(einstein, cmap="gray"), plt.axis("off");
```

# Classification

```python
face_cascade = cv2.CascadeClassifier("data/haarcascade_frontalface_default.xml")

face_rect = face_cascade.detectMultiScale(einstein)

for (x,y,w,h) in face_rect:
    
    cv2.rectangle(einstein, (x,y), (x+w, y+h), (255,255,255), 10)
    
plt.figure(), plt.title("Einstein"), plt.imshow(einstein, cmap="gray"), plt.axis("off");
```

```python
import cv2
import matplotlib.pyplot as plt

# Load Haar cascade (Make sure the path is correct)
face_cascade = cv2.CascadeClassifier("data/haarcascade_frontalface_default.xml")

# Load image
barce = cv2.imread("data/barcelona.jpg")

# Convert to grayscale (Haar cascades require grayscale images)
gray = cv2.cvtColor(barce, cv2.COLOR_BGR2GRAY)

# Face detection with optimized parameters
face_rect = face_cascade.detectMultiScale(
    gray, 
    scaleFactor=1.2,   # Increase this to avoid detecting non-face objects
    minNeighbors=10,    # Higher value = fewer false positives
    minSize=(50, 50),   # Ignore very small detections
    maxSize=(500, 500)  # Ignore extremely large detections
)

# Draw rectangles around detected faces
for (x, y, w, h) in face_rect:
    cv2.rectangle(barce, (x, y), (x + w, y + h), (255, 255, 255), 5)

# Display the result
plt.figure(figsize=(7, 7))
plt.title("Barce - Face Detection")
plt.imshow(cv2.cvtColor(barce, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

# Face Detection with Camera

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    if ret:
        
        face_rect = face_cascade.detectMultiScale(frame, minNeighbors=7)
        
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (255,255,255), 3)
        cv2.imshow("Face Detect", frame)
        
    if cv2.waitKey(1) & 0xFF == ord("q"): break
        
cap.release()
cv2.destroyAllWindows()
```
